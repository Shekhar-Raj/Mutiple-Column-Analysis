```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("BlackFriday.csv")
df.head()
```

#  comparing age and gender .

```python
sns.set(rc={'figure.figsize': (12,6)})
sns.countplot(x="Age" ,hue="Gender",data=df)
```

Here We are taking x axis as gender and hue as age.

```python
sns.set(rc={'figure.figsize': (12,6)})
sns.countplot(x="Gender",hue="Age",data=df)
```

# Here we can see that there are more number of  unmarried people of age group 26-35 .

```python
sns.set(rc={'figure.figsize': (12,6) })
sns.countplot(x="Marital_Status",hue="Age",data=df)
```

# here we can see that the Age group of 26-35 has living more in C Category Cities .

```python
sns.set(rc={'figure.figsize': (12,6)})
sns.countplot(x="Age",hue="City_Category",data=df)
```
